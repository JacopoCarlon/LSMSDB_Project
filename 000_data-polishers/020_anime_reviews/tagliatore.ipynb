```python
import sys
import os
import csv
import json
import pandas as pd
import time
import random
import pickle

#   ['username', 'anime_id', 'my_score', 'text']

filename = './cleaned_score_revs.csv'
with open(filename, 'rb') as namefile :
    thisdf = pd.read_csv(namefile)

#   result = thisdf.groupby(['username', 'anime_id']).size()

#   thisdf['sommatoria'] = thisdf.groupby(['username', 'anime_id']).agg( 'sum' )
#   thisdf.groupby(["username", "anime_id"]).count().reset_index(name="count")
newdf = thisdf.value_counts(['username', 'anime_id']).reset_index(name='count')

print(len(newdf))

uniquen = newdf.loc[newdf['count'] == 1]
print(len(uniquen))

uniquen = uniquen.drop('count', axis = 1)
print(uniquen)


merged = pd.merge(
            uniquen
        ,   thisdf
        ,   on=['username', 'anime_id']
)

merged.rename(columns={ 'username':'username' , 
                        'anime_id':'anime_id' , 
                        'score':'my_score' })

print(len(merged))
print(merged)



prudst = './pruned_anime_revs.csv'
if not os.path.exists(prudst):
    with open(prudst, 'w'): pass 
merged.to_csv(prudst, index=False)
```
