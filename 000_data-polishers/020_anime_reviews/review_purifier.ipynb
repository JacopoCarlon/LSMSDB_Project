```python
import sys
import os
import argparse
import csv
import json
import pandas as pd
import pickle

import time

t0 = time.time_ns()

already_created_dst_file = False
#   - anime : [animeID, (count_reviews), avg_reviews ] 
#   - anime : details [animeID, episodes, ... ]


to_read = 5
filename = './reviews.csv'

index_vals = ['username', 'animeid', 'score', 'text']
data = []


firaw = pd.read_csv(filename
                    #, nrows= to_read
                    )

print(f'head : {firaw.columns.values}')
print(f'num rows : {len(firaw.index)}')


t1 = time.time_ns()
print(f'time to read_src : {t1-t0} nanoseconds')


for iter in range(0, len(firaw.index)):
    first = firaw.iloc[iter]

    this_prof = first.profile
    this_anid = first.anime_uid

    stest = first.text
    #   print(f'stest type : {type(stest)} ')

    testo = stest.split('\n')
    for pit in range(0,len(testo)):
        testo[pit] = testo[pit].lstrip()
    purtesto = [el for el in testo if el != '']
    #   print(f'{first} _ type : {type(first)}')
    #   ['more pics ', 'Overall ', '8 ', 'Story ', '8 ', 'Animation ', '8 ', 'Sound ', '10 ', 'Character ', '9 ', 'Enjoyment ', '8 ', 'First things first. My "review ... #

    #   print(f'{purtesto} _ type : {type(purtesto)} _ len : {len(purtesto)}')
    ov_score = purtesto[2]
    this_scor = int(ov_score)

    #   print(f'{ov_score} _ type : {type(ov_score)} _ len : {len(ov_score)}')
    revt = purtesto[13:]

    #   print(revt[-1])
    if len(revt[-1]) < 20:
        #   print(f'removing last : {revt[-1]}')
        revt.pop(-1)
    
    this_text = revt

    data.append( [ this_prof, this_anid, this_scor, this_text ] )

    #   for sent in revt:
    #       print(sent)
    #   print(f'DBG ---> revt : {revt} _ type : {type(revt)} _ len : {len(revt)}')
    
#   print(f'DBG-> data : {data}')



################################################################



#   #   #   WORKS :
#   #   for sent in data[0][3]:
#   #       print(sent)



################################################################



#   #   CONVERT FO DATAFRAME ??
#   datafrm = pd.DataFrame(data)



################################################################



#   #   SAVE TO FILE
dst_file = './cleaned_score_revs.pkl'
#   datafrm.to_pickle(dst_file)
with open(dst_file, 'wb') as picfile :
    pickle.dump(data, picfile)
    picfile.close()


t2 = time.time_ns()
print(f'time to operate and save : {t2-t1} nanoseconds')



################################################################



#   #   #   READ FROM FILE
#   #   readerdf = pd.read_pickle(dst_file)
#   with open(dst_file, 'rb') as redfil :
#       loaded = pickle.load(redfil)
#       redfil.close()
#   
#   #   #   TEST WORKS :
#   print(f'\nDBG -> about to read from pickle file !!! :\n')
#   for sent in loaded[0][3]:
#       print(sent)


```
