```python
import sys
import os
import argparse
import csv
import json
import pandas as pd
import numpy as np

base_file = './anime-offline-database.json'

trgcols = [
        "anime_id"
    ,   "picture"
    #   ,   "thumbnail"
]

#   main_file = pd.read_json(base_file)
with open(base_file, encoding='utf-8') as trg_src :
    jobject = json.load(trg_src)

lisodict = jobject['data']

anilis = []

for oit in lisodict:
    aniname = oit["sources"]
    anipict = oit["picture"]
    #   print(f'anipict : {anipict} _ type : : {type(anipict)}')
    no_id = True
    for sit in aniname:
        if "myanimelist" in sit:
            anime_id = sit.split('/')[-1]
            no_id = False
    if no_id == False and anipict!='':
        anilis.append( [anime_id, anipict] )

newdf = pd.DataFrame(anilis, columns=trgcols)

#   SAVE TO CSV :
dst_anipictdf = './anipictdf.csv'
if not os.path.exists(dst_anipictdf):
    with open(dst_anipictdf, 'w'): pass   
newdf.to_csv(dst_anipictdf, index=False)

#   df = pd.DataFrame(data["sources" , "picture" , "thumbnail"])


#   target = main_file[trgcols]

#   print(f'exmp : {target.iloc[0]}')


```
